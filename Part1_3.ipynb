```python
import language_tool_python
tool = language_tool_python.LanguageTool('en-US')
```

```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\91880\Downloads\review_data.csv")
```

```python
df
```

```python
df=df.drop(labels=['star','app_id','reviewDate'],axis=1)
```

```python
df
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
import nltk
```

```python
nltk.download('words')
```

```python
words = set(nltk.corpus.words.words())
```

```python
for i in df.index:
    df['text'][i] = " ".join(w for w in nltk.wordpunct_tokenize(df['text'][i]) if w.lower() in words or not w.isalpha())
df
```

```python
nan=float('NaN')
df.replace("",nan,inplace=True)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df
```

```python
df['Grammar_errors']=0
df
```

```python
for i in df.index:
    text = df['text'][i]
    matches = tool.check(text)
    count = len(matches)
    if count == 0:
        df['Grammar_errors'][i]=0
    else:
        df['Grammar_errors'][i]=count
df
```

```python
df['Grammar_errors'].value_counts()
```

```python
df['Grammar_Score_(%)']=''
for i in df.index:
    df['Grammar_Score_(%)'][i]=round((df['Grammar_errors'][i]/44)*100,2)
df
```
